```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random

# Tetapkan seed untuk memastikan hasil yang konsisten
np.random.seed(42)
random.seed(42)
tf.random.set_seed(42)

# Tentukan direktori untuk data latih dan uji
train_dir = 'train animasi'
test_dir = 'test animasi'

# Preprocessing data latih tanpa augmentasi
train_datagen = ImageDataGenerator(rescale=1./255)

# Preprocessing data uji tanpa augmentasi
test_datagen = ImageDataGenerator(rescale=1./255)

# Load data latih
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(64, 64),
                                                    batch_size=32,
                                                    class_mode='categorical')

# Load data uji
test_generator = test_datagen.flow_from_directory(test_dir,
                                                  target_size=(64, 64),
                                                  batch_size=32,
                                                  class_mode='categorical',
                                                  shuffle=False)

# Membangun model CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(6, activation='softmax')  
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# Tentukan callback untuk EarlyStopping dan ModelCheckpoint
early_stopping = EarlyStopping(monitor='val_loss',  # atau 'val_accuracy'
                               patience=5,         # Jumlah epoch tanpa perbaikan setelah itu pelatihan akan dihentikan
                               restore_best_weights=True)  # Mengembalikan bobot model ke epoch dengan performa terbaik

model_checkpoint = ModelCheckpoint('best_model_animasi.keras',  # Nama file untuk menyimpan model dengan ekstensi .keras
                                   monitor='val_loss',  # Monitor berdasarkan 'val_loss'
                                   save_best_only=True)  # Hanya menyimpan model dengan performa terbaik

# Melatih model dengan EarlyStopping dan ModelCheckpoint
history = model.fit(
    train_generator,
    epochs=100,  
    validation_data=test_generator,
    callbacks=[early_stopping, model_checkpoint]  # Tambahkan callback di sini
)

# Memuat model terbaik yang telah disimpan
best_model = tf.keras.models.load_model('best_model_animasi.keras')

# Evaluasi model pada data uji
test_loss, test_acc = best_model.evaluate(test_generator)

# Membuat prediksi
predictions = best_model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Membuat DataFrame untuk menampilkan prediksi dan true labels
results_df = pd.DataFrame({'Filename': test_generator.filenames,
                           'Predicted': [class_labels[i] for i in predicted_classes],
                           'True': [class_labels[i] for i in true_classes]})

# Menampilkan seluruh tabel hasil
from IPython.display import display
display(results_df)

# Menampilkan beberapa contoh prediksi dan nilai sebenarnya
for i in range(10):  # Tampilkan 10 contoh pertama
    img = plt.imread(test_dir + '/' + results_df['Filename'].iloc[i])
    plt.imshow(img)
    plt.title(f"Predicted: {results_df['Predicted'].iloc[i]}, True: {results_df['True'].iloc[i]}")
    plt.show()
```

```python
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Evaluasi model pada data uji secara keseluruhan
test_loss, test_acc = model.evaluate(test_generator, verbose=2)

# Membuat prediksi pada data uji
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Menghitung akurasi per kelas
conf_matrix = confusion_matrix(true_classes, predicted_classes)
class_accuracies = {}
for i, label in enumerate(class_labels):
    class_acc = conf_matrix[i, i] / np.sum(conf_matrix[i]) * 100
    class_accuracies[label] = class_acc

# Menampilkan akurasi per kelas
print("Accuracy per Class in Percent:")
for label, acc in class_accuracies.items():
    print(f"{label}: {acc:.2f}%")

# Menghitung loss per kelas
class_losses = {label: 0.0 for label in class_labels}
class_counts = {label: 0 for label in class_labels}

# Iterasi melalui semua batch dalam test_generator
for i in range(len(test_generator)):
    x_batch, y_batch = test_generator[i]
    batch_loss = model.evaluate(x_batch, y_batch, verbose=0)[0]
    true_batch_classes = np.argmax(y_batch, axis=1)
    
    # Mengakumulasi loss untuk setiap kelas
    for j in range(len(true_batch_classes)):
        class_label = class_labels[true_batch_classes[j]]
        class_losses[class_label] += batch_loss
        class_counts[class_label] += 1

# Rata-rata loss per kelas
for label in class_losses:
    if class_counts[label] > 0:
        class_losses[label] /= class_counts[label]
        # Tidak mengalikan dengan 100, biarkan dalam bentuk aslinya

# Menampilkan loss per kelas
print("\nLoss per Class:")
for label, loss in class_losses.items():
    print(f"{label}: {loss:.4f}")  # Cetak dengan lebih banyak desimal untuk presisi


# Menampilkan laporan klasifikasi yang mencakup precision, recall, dan F1-score dalam persen
report = classification_report(true_classes, predicted_classes, target_names=class_labels, output_dict=True)

# Konversi ke DataFrame untuk tampilan lebih baik
report_df = pd.DataFrame(report).transpose()

# Menampilkan precision, recall, dan F1-score dalam persen
print("\nPrecision, Recall, and F1-score per Class in Percent:")
for label in class_labels:
    precision = report_df.loc[label, 'precision'] * 100
    recall = report_df.loc[label, 'recall'] * 100
    f1_score = report_df.loc[label, 'f1-score'] * 100
    print(f"{label}: Precision = {precision:.2f}%, Recall = {recall:.2f}%, F1-score = {f1_score:.2f}%")

# Menghitung dan menampilkan confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# Plot loss dan akurasi selama pelatihan
# Akurasi
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
import tracemalloc
import time
import psutil
import os
import pandas as pd

# Fungsi untuk mengukur waktu dan memori komputasi untuk setiap file
def measure_computation_time_and_memory_tracemalloc(generator, model):
    results = []
    
    for i, (batch_images, batch_labels) in enumerate(generator):
        num_files_in_batch = batch_images.shape[0]  # Jumlah file dalam batch saat ini
        
        for j in range(num_files_in_batch):
            img = batch_images[j:j+1]  # Ambil gambar individu
            
            # Mulai pengukuran memori dengan tracemalloc
            tracemalloc.start()
            
            start_time = time.time()  # Mulai timer
            prediction = model.predict(img)
            end_time = time.time()  # Akhiri timer
            
            # Ambil snapshot memori setelah prediksi
            current, peak = tracemalloc.get_traced_memory()
            tracemalloc.stop()  # Hentikan pelacakan memori
            
            # Hitung waktu komputasi
            computation_time = end_time - start_time
            memory_usage = peak - current  # Penggunaan memori puncak
            
            # Tentukan prediksi dan true label
            predicted_class = np.argmax(prediction, axis=1)[0]
            true_class = np.argmax(batch_labels[j])  # Assuming one-hot encoded labels
            
            # Hitung indeks absolut file dalam generator
            file_index = i * generator.batch_size + j
            
            # Pastikan indeks tidak melebihi jumlah file
            if file_index < len(generator.filenames):
                # Simpan hasil dalam bentuk dictionary
                results.append({
                    'Filename': generator.filenames[file_index],
                   
                    'Memory Usage (Bytes)': memory_usage
                })
        
        if i == len(generator) - 1:  # Jika sudah selesai melalui semua batch
            break
    
    return pd.DataFrame(results)

# Lakukan pengukuran waktu dan memori komputasi pada data uji
results_df = measure_computation_time_and_memory_tracemalloc(test_generator, best_model)

# Menampilkan seluruh tabel hasil
from IPython.display import display
display(results_df)

# Menampilkan statistik deskriptif dari penggunaan memori
print(results_df.describe())

# Memeriksa jumlah total file yang seharusnya diproses
print("Total files in directory:", len(test_generator.filenames))

# Memeriksa jumlah total file yang telah diproses dan ada dalam DataFrame
print("Total files processed:", len(results_df))

# Pastikan kedua angka di atas sama
if len(test_generator.filenames) == len(results_df):
    print("All files have been processed and measured.")
else:
    print("Some files may not have been processed.")

# Mencari file yang tidak terproses
processed_filenames = set(results_df['Filename'])
all_filenames = set(test_generator.filenames)

# Cari file yang tidak terproses
unprocessed_filenames = all_filenames - processed_filenames

print("Files not processed:", unprocessed_filenames)
```
