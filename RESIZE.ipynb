```python
import os
import cv2

# Directories
input_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP REAL'
output_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/HSV'

# Membuat direktori output jika belum ada
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Fungsi untuk konversi citra dari BGR ke HSV dan mempertegas warna
def convert_to_hsv_and_enhance(input_path, output_path):
    # Membaca gambar
    image = cv2.imread(input_path)
    if image is None:
        print(f"Failed to read {input_path}")
        return

    # Konversi gambar ke HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Meningkatkan saturasi dan value
    hsv_image[:,:,1] = cv2.add(hsv_image[:,:,1], 50)  # Meningkatkan saturasi
    hsv_image[:,:,2] = cv2.add(hsv_image[:,:,2], 50)  # Meningkatkan value (kecerahan)

    # Konversi kembali ke BGR
    enhanced_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)

    # Menyimpan gambar hasil konversi dan peningkatan warna
    cv2.imwrite(output_path, enhanced_image)

# Iterasi melalui semua file dalam direktori input
for filename in os.listdir(input_dir):
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename)
    convert_to_hsv_and_enhance(input_path, output_path)
    print(f"Converted {filename} to HSV, enhanced colors, and saved to {output_path}")
```

```python
import os
import cv2
import numpy as np

# Direktori input dan output
input_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/HSV'
output_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/Segmentasi Warna'

# Buat direktori output jika belum ada
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Rentang warna merah dalam RGB
lower_red_rgb = np.array([150, 0, 0])
upper_red_rgb = np.array([255, 50, 50])

# Proses setiap gambar dalam direktori input
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".bmp"):
        # Baca gambar
        image_path = os.path.join(input_dir, filename)
        image = cv2.imread(image_path)

        # Konversi ke RGB
        rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Segmentasi warna merah
        mask_rgb = cv2.inRange(rgb_image, lower_red_rgb, upper_red_rgb)
        result_rgb = cv2.bitwise_and(image, image, mask=mask_rgb)

        # Simpan hasil segmentasi
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, result_rgb)

        print(f"Segmentasi selesai untuk {filename}, disimpan di {output_path}")
```

```python
import os
import cv2
import numpy as np

# Direktori input dan output
input_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/Citra Biner'
output_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/Filling'

# Buat direktori output jika belum ada
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Kernel untuk proses filling
kernel = np.ones((5,5), np.uint8)

# Proses setiap gambar dalam direktori input
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".bmp"):
        # Baca gambar biner
        image_path = os.path.join(input_dir, filename)
        bin_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Proses filling
        filled_img = cv2.morphologyEx(bin_img, cv2.MORPH_CLOSE, kernel)

        # Simpan hasil filling
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, filled_img)

        print(f"Proses filling selesai untuk {filename}, disimpan di {output_path}")
```

```python
import os
import cv2

# Direktori input dan output
input_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/Opening'
output_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/Deteksi Tepi'

# Buat direktori output jika belum ada
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Proses setiap gambar dalam direktori input
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".bmp"):
        # Baca gambar hasil opening
        image_path = os.path.join(input_dir, filename)
        opened_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Deteksi tepi menggunakan Canny
        edge = cv2.Canny(opened_img, 80, 150)

        # Lakukan thinning pada hasil deteksi tepi
        thinned_img = cv2.ximgproc.thinning(edge, cv2.ximgproc.THINNING_GUOHALL)

        # Simpan hasil deteksi tepi dan thinning
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, thinned_img)

        print(f"Proses deteksi tepi dan thinning selesai untuk {filename}, disimpan di {output_path}")
```

```python
import os
import cv2
import numpy as np

# Direktori input dan output
input_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/Deteksi Tepi'
output_dir = 'D:/JOKIAN/UNIVERSITAS ANDALAS/M ALDI FAJAR/PP Olah/LHT'

# Buat direktori output jika belum ada
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Proses setiap gambar dalam direktori input
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".bmp"):
        # Baca gambar hasil deteksi tepi
        image_path = os.path.join(input_dir, filename)
        edge = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Salin gambar untuk digunakan dalam menggambar garis
        img_new = edge.copy()

        # Deteksi garis menggunakan Hough Line Transform
        lines = cv2.HoughLinesP(img_new, rho=1, theta=np.pi/180, threshold=3, minLineLength=0, maxLineGap=15)

        # Gambar garis pada gambar
        if lines is not None:
            for line in lines:
                x1, y1, x2, y2 = line[0]
                cv2.line(img_new, (x1, y1), (x2, y2), (255, 255, 255), 3)

        # Simpan hasil deteksi garis
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, img_new)

        print(f"Proses Line Hough Transform selesai untuk {filename}, disimpan di {output_path}")
```
